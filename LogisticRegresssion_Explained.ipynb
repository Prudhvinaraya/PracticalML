```python
import numpy as np
import pandas as pd
```

```python
#now load the data 
#load the dataset
dataset=pd.read_csv('/kaggle/input/pima-indians-diabetes-database/diabetes.csv')
df=pd.DataFrame(dataset)
df.head()
```

```python
data=df.values
X=data[:,0:8]
Y=data[:,8]
#i am going to use train_test_split method for splitting the data
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
test_size=0.33
seed=7

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=test_size,random_state=seed)
model=LogisticRegression()
model.fit(X_train,Y_train)
# pred=model.predict(X_test)
# print(pred)
acc=model.score(X_test,Y_test)
print('the accuracy of the model=',acc*100)
#dont worry about the warning its because of deprication or limit reached warning
```
